```
import matplotlib.pyplot as plt
import cv2
import numpy as np
import face_recognition
```

```
srk_img=cv2.imread('/content/srk.jpg')
srk_img=cv2.cvtColor(srk_img,cv2.COLOR_BGR2RGB)
sal_img=cv2.imread('/content/salman.jpg')
sal_img=cv2.cvtColor(sal_img,cv2.COLOR_BGR2RGB)
aki_img=cv2.imread('/content/akii.jpg')
aki_img=cv2.cvtColor(aki_img,cv2.COLOR_BGR2RGB)

```

```
srk_img_encod=face_recognition.face_encodings(srk_img)[0]
sal_img_encod=face_recognition.face_encodings(sal_img)[0]
aki_img_encod=face_recognition.face_encodings(aki_img)[0]
known_face=[srk_img_encod,sal_img_encod,aki_img_encod]
```

```
# load unknown image to know
ukn_img=cv2.imread('/content/unk1.jpg')
ukn_img=cv2.cvtColor(ukn_img,cv2.COLOR_BGR2RGB)
plt.imshow(ukn_img)
```

```
ukn_img_encod=face_recognition.face_encodings(ukn_img)
```

```
from scipy.spatial import distance
for u_img in ukn_img_encod:
  res=[]
  for k_img in known_face:
    d=distance.euclidean(u_img,k_img)
    res.append(d)
  thres=0.6
  res=np.array(res)<=thres
  name="unknown"
  if res[0]:
    name='sharukh khan'
  elif res[1]:
    name='salman khan'
  elif res[2]:
    name='akshay kumar'
  print('found {} in the photo'.format(name))        
```
