```python
td=[]
def prepare():
    for cat in cato:
        path=os.path.join(DATADIR,cat)
        cals_cato=cato.index(cat)
        for img in os.listdir(path):
            try:

                img_arry=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_arry=cv2.resize(img_arry,(img_size,img_size))
                td.append([new_arry,cals_cato])
            except:
                pass
prepare()        
     
```
